```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_file = "ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
frozen_model = 'frozen_inference_graph.pb'
```

```python
# Correct method name is 'dnn_DetectionModel' (capital D)
model = cv2.dnn_DetectionModel(frozen_model, config_file)
```

```python
classLabels = []
file_name = 'labels.txt'

# Correct way to read and split labels
with open(file_name, 'rt') as fpt:
    ClassLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print( ClassLabels)
```

```python
print(len( ClassLabels))
```

```python
model.setInputSize(320, 320)               # Set input size (width, height)
model.setInputScale(1.0/127.5)            # Scale factor (normalization)
model.setInputMean((127.5, 127.5, 127.5)) # Mean subtraction values (R,G,B)
model.setInputSwapRB(True)                # Swap Red and Blue channels
```

```python
img = cv2.imread('boy.jpg')

plt.imshow(img)
```

```python
ClassIndex, confidence, bbox = model.detect(img, confThreshold=0.5)

```

```python
print(ClassIndex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
    cv2.rectangle(img, boxes, (255, 0, 0), 2)
    cv2.putText(img, ClassLabels[ClassInd - 1], (boxes[0] + 10, boxes[1] + 40), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the video or use the webcam
cap = cv2.VideoCapture('pexes-geoge-morina-53308331.mp4')
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError('Cannot open the video')

# Font settings
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

# Create a figure for displaying frames
plt.ion()  # Turn on interactive mode
fig, ax = plt.subplots()

while True:
    ret, frame = cap.read()
    if not ret:
        break  # Exit the loop if there are no frames to read

    # Assuming 'model' is already defined and loaded
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)
    print(ClassIndex)

    if len(ClassIndex) != 0: 
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd < 80:  # Adjust this condition based on your class labels
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)
                cv2.putText(frame, ClassLabels[ClassInd - 1], (boxes[0] + 10, boxes[1] + 40), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)

    # Convert BGR to RGB for displaying with Matplotlib
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Display the frame using Matplotlib
    ax.clear()  # Clear the previous frame
    ax.imshow(frame_rgb)
    ax.axis('off')  # Hide axes
    plt.pause(0.01)  # Pause to allow the plot to update

# Release the video capture
cap.release()
plt.ioff()  # Turn off interactive mode
plt.show()  # Show the last frame
```
